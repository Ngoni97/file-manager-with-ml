```python
import fitz
import pymupdf
```

```python
def get_bookmarks(filepath):
    bookmarks = {}
    with fitz.open(filepath) as file:
        toc = file.get_toc()
        for level, title, page in toc:
            bookmarks[page] = title
    return bookmarks
```
