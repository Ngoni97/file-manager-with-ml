```python
# implementing the multithread method

from threading import Thread

def square_numbers():
    for i in range(100):
        i  * i

if __name__ == "__main__":
    threads = []
    num_threads = 10

    # create threads
    for i in range(num_threads):
        thread = Thread(target=square_numbers)
        threads.append(thread)

    # start threads
    print("process started!\n")
    for thread in threads:
        thread.start()

    # join threads: wait for them to complete
    for thread in threads:
        thread.join()

    print('end main!')
```

```python
# using concurrent processess
import threading
import concurrent.futures
import pymupdf
import time

def read_ocr(page_num, book_path):
    __text = ""
    with pymupdf.open(book_path) as file:
        page = file[page_num]
        __Text = page.get_textpage_ocr(dpi=172, full=True)
        _Text = __Text.extractTEXT()
        __text += _Text
    
    return __text

def Multithreading(page_numbers, file_name):
    
    threads = []

    # create threads
    for page_num in range(page_numbers):
        thread = Thread(target=read_ocr, args=(page_num, file_name))
        threads.append(thread)

    # start threads
    start_time = time.perf_counter()
    print("process started!\n")
    for thread in threads:
        thread.start()
    # join threads
    for thread in threads:
        thread.join()
    print("process finished!\n")
    finish_time = time.perf_counter()

    print("total time taken = {:.2d}\n".format(finish_time - start_time))
```
