```python
# -*- coding: utf-8 -*-
"""
Created on Fri Oct  4 14:31:57 2024

@author: Ngoni
"""

from tkinter import *
from tkinter import StringVar
import os
from tkinter import ttk

# pop-up window for the search tool
class SearchPopup(Frame):
    """ The pop up window to ask the description of the sorting criteria
        
    """
    def __init__(self, master):
        super(SearchPopup, self).__init__(master)
        self.__Frame_1()
        self.__Frame_2()
        self.__Frame_3()
        self.__Frame_4()
        self.__Keyboard_shortcuts()
        self.test = os.path.exists('testing_the_entry_widget.txt')
    
    # creating Frames
    def __Frame_1(self):
        """ Frame one which contains the top row of font comboboxes and their labels """
        frame_1 = Frame(self.master).pack(side=TOP,expand=True, fill='both')
        stringvar = StringVar() # for the combobox
        # add label for the font combobox
        Label(frame_1, text='Select Font').pack(side=LEFT, expand=True)
        
        # font combobox
        font_list = ['Abadi MT Condensed Light','Agency FB','Agency FB Bold', 'Algerian', 'Almanac MT',
                     'Amazone BT','Architecture','Arial','Arial Black','Arial Black Italic','Arial Bold',
                     'Arial Bold Italic','Arial Italic','Arial Narrow','Arial Narrow Bold','Arial Narrow Bold Italic',
                     'Arial Narrow Italic','Arial Rounded MT Bold','Avant Garde Book BT','Avant Garde Book Oblique BT',
                     'Avant Garde Medium BT','Avant Garde Medium Oblique BT','Bank Gothic Medium BT','Baskerville Old Face',
                     'Bauhaus 93','Beesknees ITC','Bell MT','Bell MT Bold','Bell MT Italic']
        
        self.font_combo = ttk.Combobox(frame_1, width=30,
                                       textvariable=stringvar,
                                       values=font_list).pack(side=LEFT, expand=True)
        # add label for the font size combobox
        Label(frame_1, text='Font Size').pack(side=LEFT, expand=True)
        
        size_list = ['8','9','10','11','12','13']
        
        # font-size combobox
        self.font_size = ttk.Combobox(frame_1, width=10,
                                       textvariable=stringvar,
                                       values=size_list).pack(side=LEFT, expand=True)
    
    def __Frame_2(self):
        """ Frame 2 for the text label """
        frame_2 = Frame(self.master).pack(side=TOP,expand=True, fill='both')
        # Label
        Label(frame_2, text='Sorting Description',
              font=('Lucida Calligraphy Italic',12,'bold')).pack(side=LEFT, expand=True)
        
    def __Frame_3(self):
        """ Frame two which contains the next row with the title and textbox """
        frame_3 = Frame(self.master).pack(side=TOP,expand=True, fill='both')
        # Textbox
        self.txt = Text(frame_3, height=7, wrap=WORD,
                         width=100,
                         font=('Lucida Handwriting Italic',12)).pack(side=BOTTOM, expand=True)
    
    def __Frame_4(self):
        """ Frame four which contains the buttons """
        frame_4 = Frame(self.master).pack(side=TOP,expand=True, fill='both')
        # pack the buttons at the bottom of the entry field
        btn1 = Button(frame_4, text='OK', command=self._enter).pack(side=LEFT, expand=True) # OK and proceed button
        
        btn2 = Button(frame_4, text='Clear', command=self._clear).pack(side=LEFT, expand=True) # clear the entry field button
       
        btn3 = Button(frame_4, text='Cancel', command=self._cancel).pack(side=LEFT, expand=True) # close the popup window
       
    def _enter(self, event=None): # where here the None is to allow the function to run regardless of the event is defined or not
        """ Enter and proceed the text into the Machine learning Algorithms """
        # this is mainly for debugging 
        _text = self.txt.get('1.0', END)
        # open the already existing file and append items
        if self.test == True:
            with open('testing_the_entry_widget.txt', 'r+') as file:
                file.writelines(_text)
        else:
            # if file is nonexistent then create one and wite
            with open('testing_the_entry_widget.txt', 'w+') as file:
                file.writelines(_text)
        return self.txt.get('1.0', END)
    
    def _clear(self):
        """ Clearing the Entry field """
        self.txt.delete('1.0', END)

    def _cancel(self):
        """ Cancelling the process """
        self.quit()
        
    def __Keyboard_shortcuts(self):
        """ Keyboard shortcuts """
        self.bind_all('<Return>', self._enter)
        
Root = Tk()
Root.geometry('500x250')
Root.title('Search')
Root.config(background='grey')
app = SearchPopup(Root)
Root.mainloop()

# add the option of changing the font per user preference using th Radiobutton option on the Menu bar
# 1. for the entry field
# 2. for the title
# 3. also change font size

# the selection for the font must be implemented using the OptionMenu
# add the wraptext functionality in the Text widget

# fix the layout of the subFrames which aren't displaying correctly
```

```python
from tkinter import  *
 
class WinchFrame(Frame):
    def __init__(self):
        super().__init__()
 
        #  CREATE WIDGETS:
        self.winch_lbl = Label(text="Top Right", width=15)
        self.lbl_btn_up = Label(text="UP")
        self.btn_up = Button(text="OFF", width=12, fg='green')
        self.txt = Text(height=1, width=50, bg="light blue")
        self.lbl_btn_down = Label(text="DOWN")
        self.btn_down = Button(text="OFF", width=12, fg='green')
        #  LAYOUT:
        self.winch_lbl.grid(row = 1, column = 0, padx=5, pady=5)
        self.lbl_btn_up.grid(row=1, column=1, padx=5, pady=5)
        self.btn_up.grid(row=1, column=2, padx=5, pady=5)
        self.lbl_btn_down.grid(row=3, column=1, padx=5, pady=5)
        self.btn_down.grid(row=3, column=2, padx=5, pady=5)
        self.txt.grid(row = 5, column=1, columnspan=2, padx=5, pady=5)
 
 
class WinchApp(Tk):
    def __init__(self):
 
        self.frames = {}
        self.frames["Top Right"] = WinchFrame()
        self.frames["Top Right"].grid(row=0, column=0, sticky="nsew")
 
        self.Top_Right = WinchFrame()
        self.Bot_Right = WinchFrame()
        self.Top_Right.grid(row=0, column=0, sticky='nsew')
        self.Bot_Right.grid(row=5, column=0, sticky='nsew')
 
 
if __name__ == "__main__":
    app = WinchApp()
    app.mainloop
```

```python
class OperationPanel(Frame):
    def __init__(self, app, operation):
        Frame.__init__(self, app)
        if operation == "addition":
            Label(app, text="Addition:").pack()
            self.num1 = Entry(app)
            self.num1.pack(side = LEFT)
            self.num2 = Entry(app)
            self.num2.pack(side = LEFT)
            self.disp = Entry(app)
            self.disp.pack(side = LEFT)
            Button(app, text="Reset", command=self.reset_add).pack(side = RIGHT)
            Button(app, text="Add!",  command=self.calc_add).pack(side = RIGHT)

        elif operation == "subtraction":
            Label(app, text="Subtraction:").pack()
            self.num1 = Entry(app)
            self.num1.pack(side = LEFT)
            self.num2 = Entry(app)
            self.num2.pack(side = LEFT)
            self.disp = Entry(app)
            self.disp.pack(side = LEFT)
            Button(app, text="Reset", command=self.reset_sub).pack(side = RIGHT)
            Button(app, text="Subtract!", command=self.calc_sub).pack(side = RIGHT)

        elif operation == "multiplication":
            Label(app, text="Multiplication:").pack()
            self.num1 = Entry(app)
            self.num1.pack(side = LEFT)
            self.num2 = Entry(app)
            self.num2.pack(side = LEFT)
            self.disp = Entry(app)
            self.disp.pack(side = LEFT)
            Button(app, text="Reset", command=self.reset_mul).pack(side = RIGHT)
            Button(app, text="Multiply!",  command=self.calc_mul).pack(side = RIGHT)

        elif operation == "division":
            Label(app, text="Division:").pack()
            self.num1 = Entry(app)
            self.num1.pack(side = LEFT)
            self.num2 = Entry(app)
            self.num2.pack(side = LEFT)
            self.disp = Entry(app)
            self.disp.pack(side = LEFT)
            Button(app, text="Reset", command=self.reset_div).pack(side = RIGHT)
            Button(app, text="Divide!",  command=self.calc_div).pack(side = RIGHT)
app = Tk()
app.geometry('300x150')
app.config()
OperationPanel(app, "addition").pack()
OperationPanel(app, "subtraction").pack()
OperationPanel(app, "multiplication").pack()
OperationPanel(app, "division").pack()
app.mainloop()
```
